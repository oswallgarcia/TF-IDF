```python
text1 = "Python is a 2000 made-for-TV horror movie directed by Richard \
Clabaugh. The film features several cult favorite actors, including William \
Zabka of The Karate Kid fame, Wil Wheaton, Casper Van Dien, Jenny McCarthy, \
Keith Coogan, Robert Englund (best known for his role as Freddy Krueger in the \
A Nightmare on Elm Street series of films), Dana Barron, David Bowe, and Sean \
Whalen. The film concerns a genetically engineered snake, a python, that \
escapes and unleashes itself on a small town. It includes the classic final\
girl scenario evident in films like Friday the 13th. It was filmed in Los Angeles, \
 California and Malibu, California. Python was followed by two sequels: Python \
 II (2002) and Boa vs. Python (2004), both also made-for-TV films."
 
text2 = "Python, from the Greek word (πύθων/πύθωνας), is a genus of \
nonvenomous pythons[2] found in Africa and Asia. Currently, 7 species are \
recognised.[2] A member of this genus, P. reticulatus, is among the longest \
snakes known."
 
text3 = "The Colt Python is a .357 Magnum caliber revolver formerly \
manufactured by Colt's Manufacturing Company of Hartford, Connecticut. \
It is sometimes referred to as a \"Combat Magnum\".[1] It was first introduced \
in 1955, the same year as Smith &amp; Wesson's M29 .44 Magnum. The now discontinued \
Colt Python targeted the premium revolver market segment. Some firearm \
collectors and writers such as Jeff Cooper, Ian V. Hogg, Chuck Hawks, Leroy \
Thompson, Renee Smeets and Martin Dougherty have described the Python as the \
finest production revolver ever made."

```

```python
text4 = text1+" " + text2 +" "+text3
```

```python
text4 = text4.lower()
```

```python
  nltk.download('perluniprops') # first-time use only
  nltk.download('punkt') # first-time use only
  nltk.download('nonbreaking_prefixes') # first-time use only
  nltk.download('wordnet')
  word_t = nltk.word_tokenize(text4)
  print(word_t)
```

```python
vector = []
vector2 = []
NPalabras = 0
palabra = 0
palabra2 = 0
documentos = 10
aparece = 3
aparece2 = 5

for word in word_t:
  if word == 'python' or word == 'pythons':
    vector.append(1)
    NPalabras += 1
  else:
    vector.append(0)
    NPalabras += 1

for word in word_t:
  if word == 'africa':
    vector2.append(1)
  else:
    vector2.append(0)
    

for x in vector:
  if x == 1:
    palabra +=1


for x in vector2:
  if x == 1:
    palabra2 +=1


print("Vectorizamos los datos")

  
    
print(vector)
print("palabras iguales a python: ",palabra)
print(" ")
print(vector2)
print("palabras iguales a africa: ",palabra2)
print(" ")
print("Número de palabras del texto: {} palabras".format(NPalabras))

 
```

```python
def TF(palabra,NPalabras):
  tf = palabra/NPalabras
  return tf
```

```python
def IDF(documentos,aparece):
  idf = 1 + math.log(documentos/aparece)
  return idf
```

```python
def peso(tf,idf):
  peso = tf *idf
  return peso
```

```python
print("Valores de la palabra Python en el texto:")
print("El TF es: {}".format(TF(palabra,NPalabras)))
print("El IDF es: {}".format(IDF(documentos,aparece)))
print("El Peso TF-IDF es: {}".format(TF(palabra,NPalabras),IDF(documentos,aparece)))
```

```python
print("Valores de la palabra Africa en el texto:")
print("El TF es: {}".format(TF(palabra2,NPalabras)))
print("El IDF es: {}".format(IDF(documentos,aparece2)))
print("El Peso TF-IDF es: {}".format(TF(palabra2,NPalabras),IDF(documentos,aparece2)))
```
