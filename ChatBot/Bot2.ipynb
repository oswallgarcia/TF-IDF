```python
f=open('Superman.txt','r',errors = 'ignore')

raw=f.read()

raw=raw.lower()# converts to lowercase

nltk.download('perluniprops') # first-time use only
nltk.download('punkt') # first-time use only
nltk.download('nonbreaking_prefixes') # first-time use only
nltk.download('wordnet')

sent_tokens = nltk.sent_tokenize(raw)# converts to list of sentences 
word_tokens = nltk.word_tokenize(raw)# converts to list of words

print(sent_tokens)
print(word_tokens)
```

```python
lemmer = nltk.stem.WordNetLemmatizer()
#Wordnet es usado para inglés mayormente.

def LemTokens(tokens):
 return [lemmer.lemmatize(token) for token in tokens]
remove_punct_dict = dict((ord(punct), None) for punct in string.punctuation)
def LemNormalize(text):
 return LemTokens(nltk.word_tokenize(text.lower().translate(remove_punct_dict)))
```

```python
GREETING_INPUTS = ("hola", "que onda", "buenos dias", "buenas tardes", "que hay","hey", "hello","como estas","que tal","Adios")
GREETING_RESPONSES = ["Hola", "Que onda", ":D", "Hola amigo", "Que tal", "Un gusto saludarte"]

Insultos= ["No me faltes al respeto!!","serias tan amable de elgir tus palabras","comportate","Sosiegate","cuando me hables dirigete con respeto","se educado"]
Despedida= ["Adios","nos vemos","hasta luego","chao","hasta pronto","bye","que tengas un buen dia", "que te vaya bien"]
Nombre=["Soy Olark y estoy para ayudarte","Me llamo Olark","Mi nombre es Olark","Hola soy Olark"]

def greeting(sentence):
    for word in sentence.split():
        if word.lower() in GREETING_INPUTS:
            return random.choice(GREETING_RESPONSES)

     

```

```python
def response(user_response):
  robo_response=''
  sent_tokens.append(user_response)

  TfidfVec = TfidfVectorizer(tokenizer=LemNormalize, stop_words=None)
  tfidf = TfidfVec.fit_transform(sent_tokens)
  vals = cosine_similarity(tfidf[-1], tfidf)
  idx=vals.argsort()[0][-2]
  flat = vals.flatten()
  flat.sort()
  req_tfidf = flat[-2]

  print(flat[-2])

  if(req_tfidf==0):
    robo_response=robo_response+"Lo siento, no te entendí :("
    return robo_response
  elif(req_tfidf<0.15):
    robo_response=robo_response+"No estoy tan seguro, podrias ser mas especifico"
    return robo_response

  else:
    robo_response = robo_response+sent_tokens[idx]
    return robo_response
```

```python
flag=True
print("Olark: ¡Hola! Me llamo Olark y responderé tus dudas sobre la historia de Superman, si quieres salir escribe 'salir'.")

while(flag==True):
  user_response = input()
  user_response=user_response.lower()
  if(user_response!='salir'):
    if(user_response=='gracias' or user_response=='grax' ):
      flag=True
      print("Olark: De nada.")
    elif(user_response=='como te llamas' or user_response=='me dijiste que te llamas' or user_response=='como se llama usted' or user_response=='como es tu nombre' or user_response=='su nombre'):
       NombreO = random.choice(Nombre)
       print(NombreO)
    elif(user_response=='adios'or user_response=='nos vemos'or user_response=='hasta luego' or user_response=='chao' or user_response=='hasta pronto' or user_response=='bye'):
      RespuestaD = random.choice(Despedida)
      print(RespuestaD)
    elif(user_response=='eres un idiota' or user_response=='eres un pendejo' or user_response=='eres un burro' or user_response=='eres un azno' or user_response=='eres un estupido' or user_response=='eres un tarado' or user_response=='eres un imbecil' or user_response=='eres un tarugo' or user_response=='eres un bobo' or user_response=='eres un baboso'or user_response=='eres un buey' or user_response=='eres un animal'):
      RespuestaI = random.choice(Insultos)
      print(RespuestaI)
    else:
      if(greeting(user_response)!=None):
        print("Olark: "+greeting(user_response))
      else:
        print("Olark: ",end="")
        print(response(user_response))
        sent_tokens.remove(user_response)
  else:
    flag=False
    print("Olark: ¡Adiós! Cuidate..")
```
